```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('dark_background')

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
temperature_june=session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
```

```python
#  3. Convert the June temperatures to a list.
print(temperature_june)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
temperature_june_df = pd.DataFrame(temperature_june, columns = ["date", "temperature"])
temperature_june_df.head()
```

```python
temperature_june_df.set_index(temperature_june_df['date'], inplace=True)
temperature_june_df.to_string(index=False)
temperature_june_df = temperature_june_df.sort_index()
temperature_june_df.to_string(index=False)

temperature_june_df.plot.hist(bins=12)
plt.title('June Temperature')
plt.tight_layout()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
temperature_june_df.describe()
```

```python
# Further analysis
# Calculate precipitation statistics for June
precipitation_june=session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
```

```python
print(precipitation_june)
```

```python
precipitation_june_df = pd.DataFrame(precipitation_june, columns = ["date", "precipitation"])
precipitation_june_df.head()
```

```python
precipitation_june_df.set_index(precipitation_june_df['date'], inplace=True)
precipitation_june_df.to_string(index=False)
precipitation_june_df = precipitation_june_df.sort_index()
precipitation_june_df.to_string(index=False)

precipitation_june_df.plot()
plt.title('June Precipitation')
plt.xticks(rotation=90)
plt.show()
```

```python
precipitation_june_df.describe()
```
